```python
for col in df.columns:
    if df[col].dtype == 'object':
        le = LabelEncoder()
        df[col] = le.fit_transform(df[col])
df.dtypes
```

```python
df = df.drop(['EmployeeCount','Over18','StandardHours','OverTime','MaritalStatus','BusinessTravel','TotalWorkingYears','JobLevel','YearsInCurrentRole','MonthlyIncome','Age','YearsWithCurrManager','StockOptionLevel','YearsAtCompany','JobInvolvement'],axis=1)
```

```python
df.head(10)
```

```python
df=df.drop(['MonthlyRate','HourlyRate'], axis=1)
```

```python
df.max()
```

```python
X = df.drop('Attrition',axis=1)
y = df['Attrition']
```

```python
from sklearn.model_selection import train_test_split as tts
Xtrain,Xtest,ytrain,ytest = tts(X,y,test_size=0.2) 
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(Xtrain,ytrain)
rf.score(Xtest,ytest)
```

```python
from sklearn.svm import SVC
sv = SVC()
sv.fit(Xtrain,ytrain)
sv.score(Xtest,ytest)
```

```python
from sklearn.metrics import confusion_matrix
ypred = sv.predict(Xtrain)
confusion_matrix(ytrain,ypred)
```

Data is skewed

```python
train_data = df[:1176]
train_data.head()
```

```python
maj = train_data[train_data.Attrition == 0]
min = train_data[train_data.Attrition == 1]

maj.shape,min.shape
```

```python
from sklearn.utils import resample
min_upsampled = resample(min, replace=True, n_samples = 982, random_state=42)
```

```python
min_upsampled.tail()
```

```python
df_upsampled = pd.concat([maj,min_upsampled])
```

```python
df_upsampled.Attrition.value_counts()
```

```python
df_upsampled.shape
```

```python
Xtrain_up = df_upsampled.drop(['Attrition'], axis=1)
ytrain_up = df_upsampled['Attrition']
Xtrain_up.shape, ytrain_up.shape
```

```python
rf2 = RandomForestClassifier()
rf2.fit(Xtrain_up, ytrain_up)
rf2.score(Xtest,ytest)
```

```python
ypred = rf2.predict(Xtest)
confusion_matrix(ytest,ypred)
```

```python
from sklearn.metrics import classification_report
target_names = ['class 0', 'class 1']
print(classification_report(ytest, ypred, target_names=target_names))
```
